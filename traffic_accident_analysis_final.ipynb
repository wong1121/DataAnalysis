## 가해운전자 연령층별 사고유형별  교통사고

```python
age_accident = pd.read_excel('가해운전자 연령층별 사고유형별 교통사고.xls')

## 결측치 처리
age_accident.dropna(how='all', inplace=True)
age_accident.fillna(0, inplace=True)

## 컬럼명 변경
age_accident.rename(columns={'기본계획 연령별1': '연령별', '사고유형 대분류': '사고유형'}, inplace=True)

## 타입변경
age_accident.기준년도 = age_accident.기준년도.astype('int16')
age_accident.구분값 = age_accident.구분값.astype('int32')

## 셀 정리
age_accident.drop(age_accident[age_accident.사고유형 == '합계'].index, inplace=True)
age_accident.drop(age_accident[age_accident.연령별 == '합계'].index, inplace=True)

## 인덱스 리셋
age_accident.reset_index(drop=True, inplace=True)

## csv저장
age_accident.to_csv('age_accident.csv', index=False, encoding='UTF-8')
```

## 가해운전자 연령층별 요일별 교통사고

```python
age_daily_accident = pd.read_excel('가해운전자 연령층별 요일별 교통사고.xls')

## 결측치 처리
age_daily_accident[age_daily_accident.사고요일.isna()]
age_daily_accident.dropna(how='all', inplace=True)
age_daily_accident.fillna(0, inplace=True)

## 컬럼명 변경
age_daily_accident.rename(columns={'기본계획 연령별1': '연령별'}, inplace=True)

## 타입변경
age_daily_accident.기준년도 = age_daily_accident.기준년도.astype('int16')
age_daily_accident.구분값 = age_daily_accident.구분값.astype('int32')

## 셀 정리
age_daily_accident.drop(age_daily_accident[age_daily_accident.연령별 == '합계'].index, inplace=True)
age_daily_accident.drop(age_daily_accident[age_daily_accident.사고요일 == '합계'].index, inplace=True)

## 인덱스 리셋
age_daily_accident.reset_index(drop=True, inplace=True)

## csv 저장
age_daily_accident.to_csv('age_daily_accident.csv', encoding='UTF-8', index=False)
```

## 가해운전자 연령층별 시간대별 교통사고

```python
age_bytime_accident = pd.read_excel('가해운전자 연령층별 시간대별 교통사고.xls')


## 결측치
age_bytime_accident.dropna(how='all', inplace=True)
age_bytime_accident.fillna(0, inplace=True)

## 컬럼명
age_bytime_accident.rename(columns={'기본계획 연령별1': '연령별'}, inplace=True)

## 셀 정리
age_bytime_accident.drop(age_bytime_accident[age_bytime_accident.연령별 == '합계'].index, inplace=True)
age_bytime_accident.drop(age_bytime_accident[age_bytime_accident.사고시간대 == '합계'].index, inplace=True)

## 타입 변경
age_bytime_accident.기준년도 = age_bytime_accident.기준년도.astype('int16')
age_bytime_accident.구분값 = age_bytime_accident.구분값.astype('int32')

## 인덱스 리셋
age_bytime_accident.reset_index(drop=True, inplace=True)

# csv저장
age_bytime_accident.to_csv('age_bytime_accident.csv', encoding='UTF-8', index=False)
```
